```python
# Import modules

import csv
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
# Import data set and check it out! 

data = pd.read_csv('student_logins.csv')

data.head()
```

```python
data.info()
```

```python
len(data)
```

```python
# Convert session_start_time and session_end_time to datetime objects

data.session_start_time = pd.to_datetime(data.session_start_time)
data.session_end_time = pd.to_datetime(data.session_end_time)
```

```python
# Calculate duration (in seconds) by subtracting session_start_time from session_end_time

data['duration'] = (data.session_end_time - data.session_start_time) / np.timedelta64(1, 's')
```

```python
# Listing duration to be sure conversion looks correct against original data

data[0:20]
```

Running OLS Regression with given variables.
Note, I removed account_created column because I saw a few different student IDs with different account_created dates

```python
X_1 = data[['problems_completed','student_id', 'student_previous_logins_total', 'student_previous_class_logins']].values
y_1 = data['duration'].values

X_1 = sm.add_constant(X_1, prepend=True)
results_1 = sm.OLS(y_1, X_1).fit()

results_1.summary()
```

```python
# Graph entry vs. duration

plt.scatter(data['student_id'], data['duration'])
plt.show()
```

```python
# Removing outlier data (hopefully will enhance OLS fit)

data = data[0 < data.duration]
data = data[data.duration < 20000]
```

```python
len(data)
```

Re-Running OLS with outlying removed. 

```python
# Run OLS Regression
# Remember: by convention, p-values smaller than 0.05 are taken as evidence that the population coefficient is nonzero

X_2 = data[['problems_completed','student_id', 'student_previous_logins_total', 'student_previous_class_logins']].values
y_2 = data['duration'].values

X_2 = sm.add_constant(X_2, prepend=True)
results_2 = sm.OLS(y_2, X_2).fit()

results_2.summary()
```
