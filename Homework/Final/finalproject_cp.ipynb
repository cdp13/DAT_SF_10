#### Create a Basic Restaurant Index

```python
c.execute("CREATE TABLE yelp_busindex (business_id TEXT PRIMARY KEY, name TEXT, city TEXT, \
            stars INT)")
```

```python
conn.commit()
```

```python
wanted_keys = ['business_id', 'name', 'city', 'stars']
```

```python
for i in data:
    
    if 'Restaurants' in i['categories']:
        entry_list = []
        
        for n in wanted_keys:
            entry = i.get(n)
            entry_list.append(entry)
        
        c.execute('INSERT INTO yelp_busindex VALUES (?,?,?,?)', entry_list)
        del entry_list[:]
```

```python
c.execute('SELECT * from yelp_busindex').fetchall()
```

#### Create a Table of Attributes

```python
c.execute("CREATE TABLE yelp_attributes (business_id TEXT PRIMARY KEY, city TEXT, stars INT, \
                                        price_range INT, attire TEXT, wait_service INT, \
                                        alcohol TEXT, has_tv TEXT, noise_level TEXT, wifi TEXT)")
```

```python
c.execute("DROP TABLE yelp_attributes")
```

```python
conn.commit()
```

```python
identifiers = ['business_id', 'city', 'stars']
```

```python
attributes = ['Price Range', 'Attire', 'Waiter Service', 'Alcohol', 'Has TV', 'Noise Level', 'Wi-Fi']
```

```python
for i in data:
    
    if 'Restaurants' in i['categories']:
        entry_list = []
        for x in identifiers:
            ident = i.get(x)
            entry_list.append(ident)
        
        att_list = i.get('attributes')
        
        for n in attributes:
            value = att_list.get(n)
            entry_list.append(value)
        
        c.execute('INSERT INTO yelp_attributes VALUES (?,?,?,?,?,?,?,?,?,?)', entry_list)
        del entry_list[:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(knn_att, knn_target, train_size=0.8)
```

```python
# generic cross validation function
def cross_validate(X, y, classifier, k_fold):

    # derive a set of (random) training and testing indices
    k_fold_indices = cross_validation.KFold(len(y), n_folds=k_fold,
                           indices=True, shuffle=True,
                           random_state=0)

    k_score_total = 0
    # for each training and testing slices run the classifier, and score the results
    for train_slice, test_slice in k_fold_indices :

        model = classifier(X_train, y_train)

        k_score = model.score(X_test, y_test)

        k_score_total += k_score

    # return the average accuracy
    return k_score_total/k_fold
```
